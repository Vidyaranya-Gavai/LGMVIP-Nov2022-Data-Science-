```python
# Importing Libraries

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Displaying the image

imgz = cv2.imread("F:\LGMVIP [November2022]\Task 4 - Image to Pencil Sketch with Python\IMG.jpg")
img = imgz[:,:,::-1]
plt.imshow(img)
plt.show()
```

```python
# Converting to grayscale

gray_image = cv2.cvtColor(imgz, cv2.COLOR_BGR2GRAY)
img_gray = gray_image[:,:]
plt.imshow(gray_image,cmap = 'gray')
plt.show()
```

```python
# Inverting the image

inv_img = 255 - gray_image
plt.imshow(inv_img,cmap = 'gray')
plt.show()
```

```python
# Blurring the image and inverting the blurred image

blur = cv2.GaussianBlur(inv_img,(21,21),0)
inv_blur = 255-blur
p_sketch = cv2.divide(gray_image,inv_blur,scale = 256.0)
plt.imshow(p_sketch,cmap = 'gray')
plt.show()
```
